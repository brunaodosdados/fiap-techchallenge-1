# Tratativas - dados de exportação

## Importando a base de dados

```python
exportacao_raw = pd.read_csv('https://raw.githubusercontent.com/brunaodosdados/fiap-techchallenge-1/refs/heads/main/dados/dados_brutos/exportacao_vinhos_de_mesa.csv', sep = ';')
```

```python
exportacao_raw.head()
```

## Tratamento dos dados

```python
df_exportacao = exportacao_raw.copy()
```

```python
def renomear_colunas(df):
    # Loop pelas colunas a partir da terceira (índice 2)
    for i in range(2, len(df.columns)):
        col_name = df.columns[i]
        # Se o nome da coluna termina em '.1', modificar o nome
        if col_name.endswith('.1'):
            new_col_name = "valor_" + col_name.replace(".1", "")
        else:
            new_col_name = "quantidade_" + col_name
        # Atualizando o nome da coluna
        df.rename(columns={col_name: new_col_name}, inplace=True)
    return df
```

```python
# Aplicando a função ao dataframe
df_exportacao = renomear_colunas(df_exportacao)
```

```python
df_exportacao = df_exportacao.iloc[:, [0,1] + list(range(-32,0))]
df_exportacao.head()
```

```python
def separar_dataframes(df):
    # Selecionar as duas primeiras colunas
    primeiras_colunas = df.iloc[:, :2]
    
    # Criar o dataframe "quant_exportacao" com colunas que possuem 'quantidade_' no nome
    quant_cols = [col for col in df.columns if col.startswith('quantidade_')]
    quant_exportacao = pd.concat([primeiras_colunas, df[quant_cols]], axis=1)
    
    # Criar o dataframe "valor_exportacao" com colunas que possuem 'valor_' no nome
    valor_cols = [col for col in df.columns if col.startswith('valor_')]
    valor_exportacao = pd.concat([primeiras_colunas, df[valor_cols]], axis=1)
    
    return quant_exportacao, valor_exportacao
```

```python
# Aplicar a função para separar o dataset
quant_exportacao, valor_exportacao = separar_dataframes(df_exportacao)
```

```python
quant_exportacao.info()
```

```python
valor_exportacao.info()
```

```python
quant_exportacao_melted = quant_exportacao.melt(id_vars=['Id','País'], value_vars=quant_exportacao[2:],var_name='ano',value_name='quantidade')
```

```python
quant_exportacao_melted['ano'] = quant_exportacao_melted['ano'].str[-4:]
quant_exportacao_melted['ano'] = quant_exportacao_melted['ano'].astype(int)
```

```python
quant_exportacao_melted.head()
```

```python
valor_exportacao_melted = valor_exportacao.melt(id_vars=['Id','País'], value_vars=valor_exportacao[2:],var_name='ano',value_name='valor')
```

```python
valor_exportacao_melted['ano'] = valor_exportacao_melted['ano'].str[-4:]
valor_exportacao_melted['ano'] = valor_exportacao_melted['ano'].astype(int)
```

```python
valor_exportacao_melted.head()
```

```python
quant_exportacao_melted.shape
```

```python
valor_exportacao_melted.shape
```

```python
exportacao_melted = pd.merge(quant_exportacao_melted,valor_exportacao_melted[['Id','ano','valor']],on=['Id','ano'],how='left')
```

```python
total_por_ano = exportacao_melted.groupby('ano')[['quantidade', 'valor']].sum().reset_index()
total_por_ano = total_por_ano.rename(columns={'quantidade': 'quantidade_total', 'valor': 'valor_total'})
```

```python
exportacao_melted = exportacao_melted.merge(total_por_ano, on='ano')
```

```python
exportacao_melted['quantidade_pct'] = (exportacao_melted['quantidade'] / exportacao_melted['quantidade_total']) * 100
exportacao_melted['valor_pct'] = (exportacao_melted['valor'] / exportacao_melted['valor_total']) * 100
```
